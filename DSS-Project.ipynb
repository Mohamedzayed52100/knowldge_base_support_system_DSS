```python
from experta import *
```

```python
class mix_color(KnowledgeEngine):
    @DefFacts()
    def _initial_action(self):
        print("")
        print("Hi! I am here to help you mix your colors")
        print("For that you'll have to enter your colors in the following inputs")
        print("Please enter one of the next colors:")
        print("[white , black , blue , green , red , orange , yellow , brown ]")
        print("")
        yield Fact(action="mix_color")  
        
        
    #######Take 2 color from user
    @Rule(Fact(action="mix_color") , NOT(Fact(firstColor=W())))
    def color_1(self):
        self.declare(Fact(first_color=input("Enter your color: ")))
    
    @Rule(Fact(action="mix_color") , NOT(Fact(secondColor=W())))
    def color_2(self):
        self.declare(Fact(second_color=input("Enter your color: ")))
        
    
    
    #######Rules for mixing colors
    ##if the 2 colors are the same
    @Rule(Fact(action="mix_color") , Fact(first_color="white") , Fact(second_color="white"))
    def value_1(self):
        self.declare(Fact(result_color="White"))
                                                
    @Rule(Fact(action="mix_color") , Fact(first_color="black") , Fact(second_color="black"))
    def value_2(self):
        self.declare(Fact(result_color="Black"))
        
    @Rule(Fact(action="mix_color") , Fact(first_color="red") , Fact(second_color="red"))
    def value_3(self):
        self.declare(Fact(result_color="Red"))
                                                
    @Rule(Fact(action="mix_color") , Fact(first_color="orange") , Fact(second_color="orange"))
    def value_4(self):
        self.declare(Fact(result_color="Orange"))
        
    @Rule(Fact(action="mix_color") , Fact(first_color="yellow") , Fact(second_color="yellow"))
    def value_5(self):
        self.declare(Fact(result_color="Yellow"))
                                        
    @Rule(Fact(action="mix_color") , Fact(first_color="green") , Fact(second_color="green"))
    def value_6(self):
        self.declare(Fact(result_color="Green"))
                                                
    @Rule(Fact(action="mix_color") , Fact(first_color="blue") , Fact(second_color="blue"))
    def value_7(self):
        self.declare(Fact(result_color="Blue"))
    
    @Rule(Fact(action="mix_color") , Fact(first_color="brown") , Fact(second_color="brown"))
    def value_8(self):
        self.declare(Fact(result_color="Brown"))
        
    
    ##if 2 colors are different
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="yellow") , Fact(second_color="blue")) , AND(Fact(first_color="blue") , Fact(second_color="yellow"))))
    def value_9(self):
        self.declare(Fact(result_color="Green"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="red") , Fact(second_color="blue")) , AND(Fact(first_color="blue") , Fact(second_color="red"))))
    def value_10(self):
        self.declare(Fact(result_color="Purple"))  
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="red") , Fact(second_color="yellow")) , AND(Fact(first_color="yellow") , Fact(second_color="red"))))
    def value_11(self):
        self.declare(Fact(result_color="Orange"))      
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="white") , Fact(second_color="black")) , AND(Fact(first_color="black") , Fact(second_color="white"))))
    def value_12(self):
        self.declare(Fact(result_color="Grey"))        
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="green") , Fact(second_color="red")) , AND(Fact(first_color="red") , Fact(second_color="green"))))
    def value_13(self):
        self.declare(Fact(result_color="Brown"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="blue") , Fact(second_color="white")) , AND(Fact(first_color="white") , Fact(second_color="blue"))))
    def value_14(self):
        self.declare(Fact(result_color="Cyan"))
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="white") , Fact(second_color="green")) , AND(Fact(first_color="green") , Fact(second_color="white"))))
    def value_15(self):
        self.declare(Fact(result_color="Light Green"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="orange") , Fact(second_color="red")) , AND(Fact(first_color="red") , Fact(second_color="orange"))))
    def value_16(self):
        self.declare(Fact(result_color="Vermilion"))
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="orange") , Fact(second_color="yellow")) , AND(Fact(first_color="yellow") , Fact(second_color="orange"))))
    def value_17(self):
        self.declare(Fact(result_color="Amber"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="green") , Fact(second_color="blue")) , AND(Fact(first_color="blue") , Fact(second_color="green"))))
    def value_18(self):
        self.declare(Fact(result_color="Teal"))        
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="black") , Fact(second_color="blue")) , AND(Fact(first_color="blue") , Fact(second_color="black"))))
    def value_19(self):
        self.declare(Fact(result_color="Navy"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="black") , Fact(second_color="orange")) , AND(Fact(first_color="orange") , Fact(second_color="black"))))
    def value_20(self):
        self.declare(Fact(result_color="Brown")) 
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="black") , Fact(second_color="yellow")) , AND(Fact(first_color="yellow") , Fact(second_color="black"))))
    def value_21(self):
        self.declare(Fact(result_color="Mustard")) 
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="brown") , Fact(second_color="red")) , AND(Fact(first_color="red") , Fact(second_color="brown"))))
    def value_22(self):
        self.declare(Fact(result_color="Maroon")) 
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="white") , Fact(second_color="red")) , AND(Fact(first_color="red") , Fact(second_color="white"))))
    def value_23(self):
        self.declare(Fact(result_color="Pink")) 
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="blue") , Fact(second_color="orange")) , AND(Fact(first_color="orange") , Fact(second_color="blue"))))
    def value_24(self):
        self.declare(Fact(result_color="Brown"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="white") , Fact(second_color="brown")) , AND(Fact(first_color="brown") , Fact(second_color="white"))))
    def value_25(self):
        self.declare(Fact(result_color="Beige"))
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="black") , Fact(second_color="green")) , AND(Fact(first_color="green") , Fact(second_color="black"))))
    def value_26(self):
        self.declare(Fact(result_color="Darker Green"))
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="black") , Fact(second_color="red")) , AND(Fact(first_color="red") , Fact(second_color="black"))))
    def value_27(self):
        self.declare(Fact(result_color="Darker Red"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="white") , Fact(second_color="orange")) , AND(Fact(first_color="orange") , Fact(second_color="white"))))
    def value_28(self):
        self.declare(Fact(result_color="Simon"))
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="white") , Fact(second_color="yellow")) , AND(Fact(first_color="yellow") , Fact(second_color="white"))))
    def value_29(self):
        self.declare(Fact(result_color="Light Yellow"))
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="green") , Fact(second_color="orange")) , AND(Fact(first_color="orange") , Fact(second_color="green"))))
    def value_30(self):
        self.declare(Fact(result_color="Brown"))  
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="green") , Fact(second_color="yellow")) , AND(Fact(first_color="yellow") , Fact(second_color="green"))))
    def value_31(self):
        self.declare(Fact(result_color="Green-Yellow"))  

    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="green") , Fact(second_color="brown")) , AND(Fact(first_color="brown") , Fact(second_color="green"))))
    def value_32(self):
        self.declare(Fact(result_color="Forest Green"))  

    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="blue") , Fact(second_color="brown")) , AND(Fact(first_color="brown") , Fact(second_color="blue"))))
    def value_33(self):
        self.declare(Fact(result_color="Muddy Brown"))  
     
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="orange") , Fact(second_color="brown")) , AND(Fact(first_color="brown") , Fact(second_color="orange"))))
    def value_34(self):
        self.declare(Fact(result_color="Terracotta"))      
        
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="yellow") , Fact(second_color="brown")) , AND(Fact(first_color="brown") , Fact(second_color="yellow"))))
    def value_35(self):
        self.declare(Fact(result_color="Muddy Yellow"))
    
    @Rule(Fact(action="mix_color") , OR(AND(Fact(first_color="black") , Fact(second_color="brown")) , AND(Fact(first_color="brown") , Fact(second_color="black"))))
    def value_36(self):
        self.declare(Fact(result_color="Black-Brown"))
        
        
        
        
        
        
    #######Print Result
    @Rule(Fact(action='mix_color'),Fact(result_color=MATCH.result_color))
    def color(self, result_color):
        print("")
        print("The result color is %s\n" %(result_color))

```

```python
engine = mix_color()
while(1):
    engine.reset()  
    engine.run()  
    print("Would you like to discover more colors?")
    if input() == "no":
        break;
```
